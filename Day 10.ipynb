```python
import collections
import pandas as pd
import functools
from functools import lru_cache
import sympy
import numpy as np
import copy
import itertools
import math
```

```python
with open('../inputs/input10.txt', 'r') as f:
    lines = f.readlines()
n = len(lines)
for i in range(n):
    lines[i] = lines[i].strip()
    lines[i] = list(map(int,list(lines[i])))
lines
```

```python
starting_points = []
for i in range(len(lines)):
    for j in range(len(lines[0])):
        if lines[i][j] == 0:
            starting_points.append((i,j))
starting_points
```

```python
directions = [(1,0), (0,1), (-1,0), (0,-1)]
def bfs(start):
    curr = start
    Q = []
    Q.append(start)
    Q2 = []
    counter = 1
    while len(Q)>0:
        curr = Q.pop()
        for direction in directions:
            test = (curr[0]+direction[0], curr[1]+direction[1])
            #print(test)
            if test[0]>=0 and test[0]<len(lines) and test[1]>=0 and test[1]<len(lines[0]) and lines[test[0]][test[1]]==lines[curr[0]][curr[1]]+1:
                Q2.append(test)
                #print(Q2)
        if len(Q)==0:
            #print(Q2)
            Q = Q2.copy()
            counter += 1
            if counter==10:
                return len(set(Q2))
            Q2 = []
    return 0
```

```python
ans = 0
for start in starting_points:
    ans += bfs(start)
ans
```

```python
def bfs2(start):
    curr = start
    Q = []
    Q.append([start])
    Q2 = []
    counter = 1
    while len(Q)>0:
        current = Q.pop()
        for direction in directions:
            curr = list(copy.deepcopy(current))
            test = (curr[-1][0]+direction[0], curr[-1][1]+direction[1])
            #print(test)
            if test[0]>=0 and test[0]<len(lines) and test[1]>=0 and test[1]<len(lines[0]) and lines[test[0]][test[1]]==lines[curr[-1][0]][curr[-1][1]]+1:
                curr.append(test)
                Q2.append(tuple(curr))
                #print(Q2)
        if len(Q)==0:
            #print(Q2)
            Q = Q2.copy()
            counter += 1
            if counter==10:
                #print(Q2)
                return len(set(Q2))
            Q2 = []
    return 0
```

```python
ans2 = 0
for start in starting_points:
    ans2 += bfs2(start)
ans2
```
