```python
with open('../inputs/input12.txt', 'r') as f:
    lines = f.readlines()
n = len(lines)
for i in range(n):
    lines[i] = lines[i].strip()
m = len(lines[0])
lines
```

```python
sets = []
#each distinct cluster of plots will be a "set" of points

#Traversal algorithm
directions = [(1,0), (0,1), (0,-1), (-1,0)]
visited = set()
current_set = []
neighbors = []
for i in range(n):
    for j in range(m):
        if (i,j) not in visited:
            #depth first search?
            current_set = []
            neighbors = []
            
            #current_set.append(lines[i][j])
            current_set.append((i,j))
            current_letter = lines[i][j]
            neighbors.append((i,j))
            visited.add((i,j))
            #print(current_letter)
            while len(neighbors)>0:
                current_node = neighbors.pop()
                for direction in directions:
                    new_x = direction[0] + current_node[0]
                    new_y = direction[1] + current_node[1]
                    #print(new_x,new_y)
                    if new_x>=0 and new_x<n and new_y>=0 and new_y<m and (new_x, new_y) not in visited and lines[new_x][new_y] == current_letter:
                        current_set.append((new_x,new_y))
                        visited.add((new_x,new_y))
                        neighbors.append((new_x,new_y))
            sets.append(current_set)            
                        
                    
```

```python
ans = 0
for s in sets:
    area = 0
    perimeter = 0
    for plot in s:
        area += 1
        #Check North:
        if plot[0]-1<0:
            perimeter += 1
        elif (plot[0]-1, plot[1]) not in s:
            perimeter += 1
        #Check West
        if plot[1]-1<0:
            perimeter += 1
        elif (plot[0], plot[1]-1) not in s:
            perimeter += 1
        #Check East
        if plot[1]+1==m:
            perimeter += 1
        elif (plot[0], plot[1]+1) not in s:
            perimeter += 1
        #Check South:
        if plot[0]+1==n:
            perimeter += 1
        elif (plot[0]+1, plot[1]) not in s:
            perimeter += 1
    ans += area * perimeter
ans
```

```python
#Part 2
ans = 0
for s in sets:
    area = 0
    horizontals = set()
    verticals = set()
    for plot in s:
        area += 1
        if plot[0]-1<0:
            horizontals.add((plot[0], plot[1]))
        elif (plot[0]-1, plot[1]) not in s:
            horizontals.add((plot[0], plot[1]))
        if plot[1]-1<0:
            verticals.add((plot[0], plot[1]))
        elif (plot[0], plot[1]-1) not in s:
            verticals.add((plot[0], plot[1]))
        if plot[1]+1==m:
            verticals.add((plot[0], plot[1]+1))
        elif (plot[0], plot[1]+1) not in s:
            verticals.add((plot[0], plot[1]+1))
        if plot[0]+1==n:
            horizontals.add((plot[0]+1, plot[1]))
        elif (plot[0]+1, plot[1]) not in s:
            horizontals.add((plot[0]+1, plot[1]))
    sides = 0
    for i in range(n+1):
        flag = False
        for j in range(m+1):
            if (i,j) in horizontals:
                if not flag:
                    flag = True
                    sides += 1
                if (i-1,j+1) in verticals or (i,j+1) in verticals:
                    flag = False
            else:
                flag = False
    for j in range(m+1):
        flag = False
        for i in range(n+1):
            if (i,j) in verticals:
                if not flag:
                    flag = True
                    sides += 1
                if (i+1,j-1) in horizontals or (i+1,j) in horizontals:
                    flag = False
            else:
                flag = False
    ans += area * sides
ans
```
