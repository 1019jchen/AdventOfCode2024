```python
import collections
import pandas as pd
import functools
from functools import lru_cache
import sympy
import numpy as np
import copy
import itertools
import re
import math
from sympy import *
from heapq import heapify, heappush, heappop 
```

```python
with open('../inputs/input16.txt', 'r') as f:
    lines = f.readlines()
n = len(lines)
sx = 0
sy = 0
ex = 0
ey = 0
direct = 0
directs = [(0,1), (1,0), (0,-1), (-1,0)]
for i in range(n):
    lines[i] = lines[i].strip()
    if "S" in lines[i]:
        sx = i
        sy = lines[i].index("S")
    if "E" in lines[i]:
        ex = i
        ey = lines[i].index("E")
print(sx,sy,ex,ey)
m = len(lines[0])
lines
```

```python
visited = set()
low_score = float('inf')
visited.add((sx,sy, direct))
low_paths = set()
#score, coords, direction, visited
Q = [(0,sx,sy, direct,[(sx,sy,direct)])]
memo = {}
heapify(Q)
while Q:
    node = heappop(Q)
    if (node[1],node[2], node[3]) not in memo:
        memo[(node[1],node[2],node[3])] = node[0]
    if lines[node[1]][node[2]]=="E":
        if node[0]<=low_score:
            low_score = node[0]
            print(node[0])
            for seat in node[4]:
                low_paths.add(seat)
    dx = directs[node[3]][0]
    dy = directs[node[3]][1]
    if lines[node[1]+dx][node[2]+dy] != "#":
        if (node[1]+dx,node[2]+dy,node[3]) not in visited:
            temp = list(node[4])
            temp.append((node[1]+dx,node[2]+dy, node[3],node[0]+1))
            new_path = tuple(temp)
            heappush(Q,(node[0]+1,node[1]+dx,node[2]+dy,node[3], new_path))
            visited.add((node[1]+dx,node[2]+dy,node[3]))    
    left90 = (node[3]-1)%4
    dx = (directs[left90][0])
    dy = directs[left90][1]
    if lines[node[1]+dx][node[2]+dy] != "#":
        if (node[1]+dx,node[2]+dy,left90) not in visited:
            temp = list(node[4])
            temp.append((node[1]+dx,node[2]+dy,left90,node[0]+1001))
            new_path = tuple(temp)
            heappush(Q,(node[0]+1001,node[1]+dx, node[2]+dy, left90,new_path))
            visited.add((node[1]+dx,node[2]+dy,left90))
    right90 = (node[3]+1)%4
    dx = (directs[right90][0])
    dy = directs[right90][1]
    if lines[node[1]+dx][node[2]+dy] != "#":
        if (node[1]+dx,node[2]+dy,right90) not in visited:
            temp = list(node[4])
            temp.append((node[1]+dx,node[2]+dy,right90,node[0]+1001))
            new_path = tuple(temp)
            heappush(Q,(node[0]+1001,node[1]+dx, node[2]+dy, right90,new_path))
            visited.add((node[1]+dx,node[2]+dy, right90))
```
