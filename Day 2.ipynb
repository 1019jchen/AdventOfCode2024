```python
import collections
import pandas as pd
import functools
from functools import lru_cache
import sympy
import numpy as np
```

```python
with open('../inputs/input2.txt', 'r') as f:
    lines = f.readlines()
n = len(lines)
for i in range(n):
    lines[i] = lines[i].strip()
    lines[i] = lines[i].split(" ")
lines
```

```python
for i in range(n):
    for num in range(len(lines[i])):
        lines[i][num] = int(lines[i][num])
        
lines
```

```python
safe = 0
increasing = [1,2,3]
decreasing = [-1,-2,-3]
for report in range(len(lines)):
    diffs = []
    for x in range(1,len(lines[report])):
        diff = lines[report][x]-lines[report][x-1]
        diffs.append(diff)
    condensed = collections.Counter(diffs)
    seen_increasing = 0
    seen_decreasing = 0
    out_of_range = 0
    for c in condensed:
        if c in increasing:
            seen_increasing += 1
        elif c in decreasing:
            seen_decreasing += 1
        else:
            out_of_range += 1
    if out_of_range == 0 and( (seen_increasing==0 and seen_decreasing>0) or (seen_increasing>0 and seen_decreasing ==0)):
        safe+= 1
safe
```

```python
def detect_safe(report):
    diffs = []
    for x in range(1,len(report)):
        diff = report[x]-report[x-1]
        diffs.append(diff)
    condensed = collections.Counter(diffs)
    seen_increasing = 0
    seen_decreasing = 0
    out_of_range = 0
    for c in condensed:
        if c in increasing:
            seen_increasing += 1
        elif c in decreasing:
            seen_decreasing += 1
        else:
            out_of_range += 1
    if out_of_range == 0 and( (seen_increasing==0 and seen_decreasing>0) or (seen_increasing>0 and seen_decreasing ==0)):
        return True
```

```python
safe = 0
for report in range(len(lines)):
    report_done = False
    for i in range(len(lines[report])):
        temp = lines[report].copy()
        temp.pop(i)
        #print(temp)
        if detect_safe(temp) and not report_done:
            safe += 1
            report_done = True
safe
```
