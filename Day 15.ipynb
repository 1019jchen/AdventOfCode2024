```python
def move2(i):
    d = dirs[instructions[i]]
    xnew = d[0]+cur_x
    ynew = d[1]+cur_y
    #print(xnew,ynew)
    if graph[xnew][ynew] == '#':
        #print(1)
        return cur_x, cur_y
    elif graph[xnew][ynew] == '.':
        #print(2)
        graph[cur_x][cur_y] = '.'
        graph[xnew][ynew] = '@'
        return xnew, ynew
    elif graph[xnew][ynew] == ']':
        movable_o = []
        movable_o.append((cur_x,cur_y))
        #print(xnew,ynew)
        rightx = xnew
        righty = ynew
        leftx = xnew
        lefty = ynew-1
        new_boxes = []
        new_boxes.append((rightx,righty))
        new_boxes.append((leftx,lefty))
        while new_boxes:
            #print("NB",new_boxes)
            ox,oy = new_boxes.pop(0)
            #print(ox,oy)
            if graph[ox+d[0]][oy+d[1]] == '#':
                return cur_x, cur_y
            elif graph[ox+d[0]][oy+d[1]] == ']' and (ox+d[0],oy+d[1]) not in new_boxes and (ox+d[0],oy+d[1]) not in movable_o:
                #print("Append1", (ox+d[0],oy+d[1]))
                new_boxes.append((ox+d[0],oy+d[1]))
                new_boxes.append((ox+d[0],oy+d[1]-1))
            elif graph[ox+d[0]][oy+d[1]] == '[' and (ox+d[0],oy+d[1]) not in new_boxes and (ox+d[0],oy+d[1]) not in movable_o:
                #print("Append2", (ox+d[0],oy+d[1]))
                new_boxes.append((ox+d[0],oy+d[1]+1))
                new_boxes.append((ox+d[0],oy+d[1]))
            movable_o.append((ox,oy))
            #print("MO",movable_o)
        for a in range(len(movable_o)-1,-1,-1):
            #print(graph[movable_o[a][0]+d[0]][movable_o[a][1]+d[1]],graph[movable_o[a][0]][movable_o[a][1]])
            graph[movable_o[a][0]+d[0]][movable_o[a][1]+d[1]] = graph[movable_o[a][0]][movable_o[a][1]]
            graph[movable_o[a][0]][movable_o[a][1]] = '.'
        return xnew,ynew
    elif graph[xnew][ynew] == '[':
        movable_o = []
        movable_o.append((cur_x,cur_y))
        #print(xnew,ynew)
        leftx = xnew
        lefty = ynew
        rightx = xnew
        righty = ynew+1
        new_boxes = []
        new_boxes.append((leftx,lefty))
        new_boxes.append((rightx,righty))
        while new_boxes:
            #print("NB",new_boxes)
            ox,oy = new_boxes.pop(0)
            #print(ox,oy)
            if graph[ox+d[0]][oy+d[1]] == '#':
                return cur_x, cur_y
            elif graph[ox+d[0]][oy+d[1]] == ']' and (ox+d[0],oy+d[1]) not in new_boxes and (ox+d[0],oy+d[1]) not in movable_o:
                #print("Append1", (ox+d[0],oy+d[1]))
                new_boxes.append((ox+d[0],oy+d[1]-1))
                new_boxes.append((ox+d[0],oy+d[1]))
            elif graph[ox+d[0]][oy+d[1]] == '[' and (ox+d[0],oy+d[1]) not in new_boxes and (ox+d[0],oy+d[1]) not in movable_o:
                #print("Append2", (ox+d[0],oy+d[1]))
                new_boxes.append((ox+d[0],oy+d[1]))
                new_boxes.append((ox+d[0],oy+d[1]+1))
            movable_o.append((ox,oy))
            #print("MO",movable_o)
        for a in range(len(movable_o)-1,-1,-1):
            #print(graph[movable_o[a][0]+d[0]][movable_o[a][1]+d[1]],graph[movable_o[a][0]][movable_o[a][1]])
            graph[movable_o[a][0]+d[0]][movable_o[a][1]+d[1]] = graph[movable_o[a][0]][movable_o[a][1]]
            graph[movable_o[a][0]][movable_o[a][1]] = '.'
        return xnew,ynew
```
