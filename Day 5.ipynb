```python
import collections
import pandas as pd
import functools
from functools import lru_cache
import sympy
import numpy as np
```

```python
with open('../inputs/input5.txt', 'r') as f:
    lines = f.readlines()
n = len(lines)
for i in range(n):
    lines[i] = lines[i].strip()
    lines[i] = list(map(int, lines[i].split("|")))
lines
```

```python
with open('../inputs/input5pages.txt', 'r') as f:
    pages = f.readlines()
n = len(pages)
for i in range(n):
    pages[i] = pages[i].strip()
    pages[i] = list(map(int, pages[i].split(",")))
pages
```

```python
count = 0
incorrect_pages = []
for page in pages:
    possible = True
    for line in lines:
        if line[0] in page and line[1] in page and page.index(line[0]) > page.index(line[1]):
            print(page, line[0], line[1])
            possible = False
            break
    if possible:
        count += page[len(page)//2]
    else:
        incorrect_pages.append(page)
count
```

```python
rules = collections.defaultdict(list)
for line in lines:
    rules[line[0]].append(line[1])
rules
```

```python
for x,page_ in enumerate(incorrect_pages):
    page = page_.copy()
    for i in range(len(page)):
        for j in range(i+1,len(page)):
            #print(page,i,j)
            if page[i] in rules[page[j]]:
                page[j], page[i] = page[i], page[j]
    incorrect_pages[x] = page
count2 = 0
incorrect_pages
```
