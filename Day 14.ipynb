```python
import collections
import pandas as pd
import functools
from functools import lru_cache
import sympy
import numpy as np
import copy
import itertools
import re
import math
from sympy import *
```

```python
with open('../inputs/input14.txt', 'r') as f:
    lines = f.readlines()
n = len(lines)
for i in range(n):
    lines[i] = lines[i].strip().replace("=",",").replace(" ",",").split(",")
m = len(lines[0])
dimensions_x = 101
dimensions_y = 103
iterations = 100
```

```python
robots = []
for robot in lines:
    robots.append((int(robot[1]), int(robot[2]), int(robot[4]), int(robot[5])))
```

```python
new_robots = []
for i,robot in enumerate(robots):
    new_x = (robot[0]+iterations*robot[2])%dimensions_x
    new_y = (robot[1]+iterations*robot[3])%dimensions_y
    new_robots.append((new_x,new_y))
```

```python
q1 = 0
q2 = 0
q3 = 0
q4 = 0
for r in new_robots:
    if r[0]<dimensions_x//2:
        if r[1]<dimensions_y//2:
            q1+=1
        elif r[1]>dimensions_y//2:
            q3 += 1
    elif r[0]>dimensions_x//2:
        if r[1]>dimensions_y//2:
            q4+=1
        elif r[1]<dimensions_y//2:
            q2+=1
print(q1,q2,q3,q4)
```

```python
ans = q1*q2*q3*q4
ans
```
